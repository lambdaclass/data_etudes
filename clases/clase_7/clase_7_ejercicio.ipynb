*CIFAR-10* es un dataset similar al *MNIST*, consiste de imágenes de 32x32 pixels y 10 categorías distintas, solo que estas categorías ahora son:
- Avión
- Auto
- Pájaro
- Gato
- Ciervo
- Perro
- Rana
- Caballo
- Barco
- Camión

https://www.cs.toronto.edu/~kriz/cifar.html

Probar distintas redes convolucionales sobre el dataset.

```python
import matplotlib.pyplot as plt
from keras.datasets import cifar10
```

```python
(trainX, trainy), (testX, testy) = cifar10.load_data()
```

```python
print('Train: X=%s, y=%s' % (trainX.shape, trainy.shape))
print('Test: X=%s, y=%s' % (testX.shape, testy.shape))
```

```python
for i in range(9):
    plt.subplot(330 + 1 + i)
    plt.imshow(trainX[i])
plt.show()
```
